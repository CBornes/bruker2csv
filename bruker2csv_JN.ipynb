```python
# Import required packages
import os
import nmrglue as ng
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the path to the file you want to export below (ex: C/Bruker/TopSpin4.1.3/data/sample/pdata/1)
data_dir = os.path.normpath('C:/Bruker/TopSpin4.1.3/data/400/Zeolite/22/pdata/1')
dic, data = ng.bruker.read_pdata(data_dir)
udic = ng.bruker.guess_udic(dic, data)
uc = ng.fileiobase.uc_from_udic(udic)
ppm_scale = uc.ppm_scale()
norm_data = data/data.max()
df = pd.DataFrame()
df['ppm'] = ppm_scale
df['int'] = data
df['norm int'] = norm_data
```

```python
# Plot unormalized data
plt.plot(ppm_scale, data)
plt.xlim(20,-10)
plt.xlabel('$^1$H (ppm)')
```

```python
# Plot ormalized data
plt.plot(ppm_scale, norm_data)
plt.xlim(20,-10)
plt.xlabel('$^1$H (ppm)')
```
